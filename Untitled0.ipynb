```python
import json
import requests
import matplotlib.pyplot as plt
from datetime import datetime

# Визначаємо період: з 1 по 8 травня 2025 року
start_date = "20250501"
end_date = "20250508"

# Формуємо URL для запиту до API НБУ
url = f"https://bank.gov.ua/NBU_Exchange/exchange_site?start={start_date}&end={end_date}&valcode=eur&json"

# Виконуємо GET-запит до API
response = requests.get(url)
data = json.loads(response.content)

# Ініціалізуємо списки для дат та курсів
dates = []
rates = []

# Обробляємо отримані дані
for item in data:
    date_str = item['exchangedate']
    rate = item['rate']
    # Перетворюємо рядок дати у формат datetime
    date_obj = datetime.strptime(date_str, "%d.%m.%Y")
    dates.append(date_obj)
    rates.append(rate)
    # Виводимо дату та курс
    print(f"{date_str}: {rate} UAH")

# Будуємо графік зміни курсу
plt.figure(figsize=(10, 5))
plt.plot(dates, rates, marker='o', linestyle='-', color='blue')
plt.title("Зміна офіційного курсу євро (НБУ) за період 01.05.2025 - 08.05.2025")
plt.xlabel("Дата")
plt.ylabel("Курс EUR/UAH")
plt.grid(True)
plt.tight_layout()
plt.show()
```
